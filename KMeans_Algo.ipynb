```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import plotly as py
import plotly.graph_objs as go
from sklearn.cluster import KMeans
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Mall_Customers.csv")
```

```python
data
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
x = data.iloc[:,[3,4]].values
x
```

```python

clust = []

for i in range(1,11):
  kmeans = KMeans(n_clusters= i, init = "k-means++", random_state=0)
  kmeans.fit(x)
  clust.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),clust)
plt.title('The Elbow Method')
plt.xlabel('No.of clusters')
plt.ylabel('clust')
plt.show()
```

```python
model = KMeans(n_clusters= i, init = "k-means++", random_state=0)
y_kmeans = model.fit_predict(x)
```

```python

plt.scatter(x[y_kmeans==0,0],x[y_kmeans==0,1],s=30,c='red',label='cluster 1',marker="o")
plt.scatter(x[y_kmeans==1,0],x[y_kmeans==1,1],s=30,c='blue',label='cluster 2',marker="o")
plt.scatter(x[y_kmeans==2,0],x[y_kmeans==2,1],s=30,c='green',label='cluster 3',marker="o")
plt.scatter(x[y_kmeans==3,0],x[y_kmeans==3,1],s=30,c='yellow',label='cluster 4',marker="o")
plt.scatter(x[y_kmeans==4,0],x[y_kmeans==4,1],s=30,c='orange',label='cluster 5',marker="o")
plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1],s=60,c='black',marker ='*',label='centroids')
plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
